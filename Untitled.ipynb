```python
def page2table(table: str, code: str=None):
    """抽取行业配置tabel"""
    datadict = {}
    for tr in [r for r in table.split('\n') if len(r) > 1]:
        arr=tr.strip().split()
        datadict[arr[1]] = float(arr[2].replace('%', ''))
    return datadict
```

```python
def df2dict(df, main='code'):
    """将数据框转成以main对应的列作为key的字典
    """
    d = {}
    mat = df.drop([main],axis=1).values.tolist()
    colname = df.drop([main],axis=1).columns.values.tolist()
    for i, c in enumerate(df[main].values.tolist()):
        c = str(c)
        if len(c) < 6:
            c = '0' * (6-len(c)) + c
        d[c] = {}
        for j, n in enumerate(colname):
            d[c][n] = mat[i][j]
    return d
dct = df2dict(df)
```

```python
#weeks = 0
def invest(weeks=0):
    """计算当前投资在每个行业（industry）
    """
    tdict = dct
    fdict = ddict
    idict = {i: 0 for i in industry}

    for cd in tdict.keys():
        for bn in idict.keys():
            plus = 0 if weeks == 0 or tdict[cd]['period'] == 0 else weeks // \
                tdict[cd]['period'] * tdict[cd]['turn']
            if bn in fdict[cd]:
                idict[bn] += (tdict[cd]['money'] + plus) * fdict[cd][bn] * 0.01
    return idict

idict
```
